```python
continent_pop = df.groupby("continent")["2023 population"].sum()
continent_pop.plot(kind = "pie", autopct= '%1.1f%%')
plt.title("Population Distribution by Continent(2023)")
plt.show()

print("===INSIGHTS===")
print("\n1. We can see the asia accommodates the largest portion of the population, considering it is also the largest continent.")
print("\n2. Whereas Oceania only accommodates about 0.6% of golbal population. Knowing the wilderness of Australia its reasonable.")
```

```python
df["Density_1970"] = df["1970 population"] / df["area (km²)"]
df["Density_2023"] = df["2023 population"] / df["area (km²)"]
```

```python
import matplotlib.pyplot as plt
import numpy as np

top10 = df.sort_values("2023 population", ascending=False).head(10)

x = np.arange(len(top10))
width = 0.35

plt.figure()

plt.bar(x - width/2, top10["Density_1970"], width, label="1970")
plt.bar(x + width/2, top10["Density_2023"], width, label="2023")

plt.xticks(x, top10["country"], rotation=45)
plt.title("Population Density: 1970 vs 2023 (Top 10 Countries)")
plt.legend()

plt.show()

print("===INSIGHTS===")
print("\n1. Bangladesh stands out clearly — its density has increased from already high (~450) to extremely high (~1170+), far exceeding all other countries. This indicates intense population pressure on limited land, making it one of the most densely populated large countries in the world.")
print("\n2. China’s density has increased, but not as dramatically as India, indicating more controlled population growth, possibly due to historical population policies.")
print("\n3. Countries like the United States and Russia have low density values and only slight increases over time. This suggests slower population growth and more land availability, resulting in less pressure on space.")
```

```python
import matplotlib.pyplot as plt

colors = {
    "Asia": "red",
    "Europe": "blue",
    "Africa": "green",
    "North America": "orange",
    "South America": "purple",
    "Oceania": "brown"
}

plt.figure()

for continent in df["continent"].unique():
    subset = df[df["continent"] == continent]
    
    plt.scatter(
        subset["area (km²)"],
        subset["2023 population"],
        s=subset["density (km²)"] / 2,
        alpha=0.6,
        color=colors.get(continent, "gray"),
        label=continent
    )

plt.xscale("log")
plt.yscale("log")

plt.xlabel("Area (log scale)")
plt.ylabel("Population (log scale)")
plt.title("Area vs Population (Colored by Continent)")

plt.legend(loc="lower right")
plt.show()

print("===INSIGHTS===")
print("\n1. While larger countries generally tend to have higher populations, the relationship is not perfectly linear. Some mid-sized countries appear higher on the population axis than larger ones, showing that factors beyond land size (like resources, economy, and history) strongly influence population distribution.")
print("\n2. Even within a single continent, bubbles (density) vary significantly in size, indicating that countries in the same region can have completely different population pressures, depending on urbanization, geography, and development patterns.")
print("\n3. Several countries with smaller area (left side of the plot) have relatively large bubble sizes, indicating high population density despite limited land.")
```

```python
plt.figure()

sns.boxplot(x="continent", y="density (km²)", data=df)
sns.stripplot(x="continent", y="density (km²)", data=df,
              color="black", alpha=0.3)

plt.yscale("log")

plt.title("Population Density Distribution by Continent")
plt.xticks(rotation=30)

plt.show()
print("===INSIGHTS===")
print("\n1. While a pie chart shows the proportion of total population by continent, it does not reveal how population is distributed within each continent.")
print("\n2. Despite the presence of extreme outliers, the median values for all continents lie in a relatively low range, meaning most countries are not densely populated, this indicates that global population pressure is concentrated in a few countries, rather than being evenly distributed worldwide.")
print("\n3. The boxplot shows that continents like Asia and Europe have a wider spread and multiple extreme outliers, indicating that population density varies significantly within these regions.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df["CAGR"] = ((df["2023 population"] / df["1970 population"]) ** (1/53)) - 1

top_cagr = df.sort_values("CAGR", ascending=False).head(10)

plt.figure()

sns.barplot(x="CAGR", y="country", data=top_cagr)

plt.title("Top 10 Countries by Population Growth Rate (CAGR)")
plt.xlabel("CAGR")
plt.ylabel("Country")

plt.show()
print("===INSIGHTS===")
print("\n1. CAGR represents the average annual growth rate over time, allowing comparison of population growth across countries regardless of their initial size.")
print("\n2. Countries with the highest CAGR are not the most populous ones, indicating that smaller or developing nations tend to grow faster in percentage terms.")

print("\n3. High CAGR values are mostly observed in developing regions, suggesting that future population growth will be concentrated in these areas rather than already large, developed countries.")
```
